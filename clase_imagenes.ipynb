```python
import cv2
from matplotlib import pyplot as plt

image = cv2.imread('IMG_2006.jpg', cv2.IMREAD_COLOR)

plt.figure()

plt.subplot(141)
plt.title('original')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))

plt.subplot(142)
plt.title('canal R')
plt.imshow(image[:,:,0], cmap='Reds')  # Cambiado el cmap a 'Reds'

plt.subplot(143)
plt.title('canal G')
plt.imshow(image[:,:,1], cmap='Greens')  # Cambiado el cmap a 'Greens'

plt.subplot(144)
plt.title('canal B')
plt.imshow(image[:,:,2], cmap='Blues')  # Cambiado el cmap a 'Blues'

plt.show()

```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('IMG_2006.jpg', cv2.IMREAD_GRAYSCALE)

original_img = image.dtype
print('Profundidad original de color:', original_img)

image_16bits = np.uint16(image)
nueva_img = image_16bits.dtype
print('Profundidad de color despu√©s de convertir a 16 bits:', nueva_img)

cv2.imwrite('image16bits.jpg', image_16bits)
img_guardada = cv2.imread('image16bits.jpg', cv2.IMREAD_UNCHANGED)

plt.figure(figsize=(10, 5))

plt.subplot(121)
plt.title('Original')
plt.imshow(image, cmap='gray')

plt.subplot(122)
plt.title('Imagen tratada 16 bits')
plt.imshow(img_guardada, cmap='gray')  # Corregido aqu√≠

plt.show()
```
