```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def solarizar(image, umbral):
    img_solarizada = np.where(image < umbral, image, 255 - image)
    return img_solarizada

def posterizar(image, nivel):
    f_nvl = 255 / nivel
    img_poster = np.round(image / f_nvl) * f_nvl
    return img_poster.astype(np.uint8)

imagen = cv2.imread('Ricoauto.jpg')
img_gris = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)

# Solarizar la imagen
umbral_solarizacion = 127
img_sol = solarizar(img_gris, umbral_solarizacion)

# Posterizar la imagen
nivel_posterizacion = 4
img_posterizada = posterizar(img_gris, nivel_posterizacion)

# Mostrar las imágenes
plt.figure(figsize=(10, 5))

plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB))
plt.title('Imagen Original')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(img_sol, cmap='gray')
plt.title('Imagen Solarizada (Umbral=127)')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(img_posterizada, cmap='gray')
plt.title('Imagen Posterizada (Nivel=4)')
plt.axis('off')

plt.show()
```

```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize']=[10,5]
tamanio_img = 100
R = np.ones((tamanio_img, tamanio_img)) * 255
G = np.ones((tamanio_img, tamanio_img)) * 128
B = np.ones((tamanio_img, tamanio_img)) * 0

imagen = np.dstack((R, G, B))
imagen = np.array(imagen, dtype=int)

plt.figure()
plt.imshow(imagen)
plt.show()
```

```python
import cv2 
import matplotlib.pyplot as plt

# Leer la imagen original a color
img_color = cv2.imread('Cars.jpg')


img_gray = cv2.imread('Cars.jpg', cv2.IMREAD_GRAYSCALE)


negativo = 255 - img_gray


plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img_color, cv2.COLOR_BGR2RGB))
plt.title('Imagen Original (Color)')


plt.subplot(1, 2, 2)
plt.imshow(negativo, cmap='gray')
plt.title('Negativo de la Imagen (Escala de Grises)')

plt.show()
```
