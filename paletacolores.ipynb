```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Cargar la imagen
imagen = cv2.imread('cancer.jpg')

# Seleccionar la regin de inters (ROI)
celula_cancer = imagen[50:60,20:35,:]

# Guardar la ROI como una imagen separada
plt.imsave("ImagenCelularCancer.jpg", celula_cancer)

# Leer la imagen guardada
celula = plt.imread("ImagenCelularCancer.jpg")

# Calcular y mostrar el histograma por canal de color
color = ('r', 'g', 'b')
for canal, col in enumerate(color):
    histograma = cv2.calcHist([celula*255],[canal],None,[256], [0,256])
    plt.title("Histograma")
    plt.plot(histograma, color = col)
    plt.xlim([0,256])
plt.show()

```

```python
r_celula = celula[:,:,0]*255
g_celula = celula[:,:,1]*255
b_celula = celula[:,:,2]*255

r_celula = r_celula>200
g_celula = g_celula>200
b_celula = b_celula>200

img_fin = np.array(r_celula * g_celula * b_celula * 255)
plt.imshow(img_fin)
plt.show()
```

```python
import cv2
import numpy as np

def solarizar(image, umbral):
    # Solarizar la imagen utilizando la funcin de umbralizacin
    _, imagen_umbralizada = cv2.threshold(image, umbral, 255, cv2.THRESH_BINARY_INV)
    
    # Invertir la imagen umbralizada para obtener la solarizacin
    imagen_solarizada = 255 - imagen_umbralizada
    
    return imagen_solarizada
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Lista de colores en formato BGR
colores_bgr = [
    (255, 255, 255), (128, 128, 128), (211, 211, 211), (169, 169, 169), 
    (0, 0, 255), (0, 128, 0), (255, 0, 0), (0, 255, 255), (255, 255, 0), 
    (255, 0, 255), (165, 42, 42), (128, 128, 0), (0, 165, 255), (255, 192, 203), (128, 0, 128)
]

# Crear la paleta de colores
paleta_colores = np.zeros((100, len(colores_bgr) * 100, 3), dtype=np.uint8)
for i, color in enumerate(colores_bgr):
    paleta_colores[:, i*100:(i+1)*100] = color

# Mostrar la paleta de colores
plt.figure(figsize=(10, 5))
plt.imshow(paleta_colores)
plt.axis('off')
plt.title('Paleta de Colores')
plt.show()

# Cargar la imagen "Motorica.jpg"
imagen_motorica = cv2.imread('Motorica.jpg')

# Aplicar la paleta de colores a la imagen
plt.figure(figsize=(15, 15))
for i, color in enumerate(colores_bgr):
    imagen_colorida = imagen_motorica.copy()
    for j in range(3):  # Para cada canal de color (B, G, R)
        imagen_colorida[:, :, j] = np.where(imagen_colorida[:, :, j] > 200, color[j], imagen_colorida[:, :, j])
    plt.subplot(5, 3, i + 1)
    plt.imshow(cv2.cvtColor(imagen_colorida, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.title(f'Color {i+1}')

plt.show()


```
